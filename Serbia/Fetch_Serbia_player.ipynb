```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

years = list(range(2011, 2023))

#url_start = 'https://basketball.realgm.com/international/league/12/French-Jeep-Elite/stats/{year}/Per_48/Qualified/All/points/{position}/desc/1/Regular_Season'
url_start = 'https://basketball.realgm.com/international/league/13/Serbian-KLS/players/|{}/{}'
data = []  # List to store parsed data
index = ['667','771','872','961','1055']

for index, year in zip(index, years):
    url = url_start.format(index, year)  # Properly formatted URL
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    rows = soup.select('.tablesaw tbody tr')
    print(f"Year: {year}, Rows: {len(rows)}")  # Debugging print

    for row in rows:
        player = [year]  # Add the year and position to the player data
        player += [cell.text.strip() for cell in row.find_all('td')]
        data.append(player)

columns = ['Year'] + [header.text.strip() for header in soup.select('.tablesaw thead th')]

print("Columns:", columns)
print("Number of columns:", len(columns))

# Creating DataFrame from the collected data
df_serbia_players = pd.DataFrame(data, columns=columns)

# Creating DataFrame from the collected data
#columns = ['Year', 'Position', '#', 'Player', 'Team', 'GP', 'MPG', 'PPG', 'FGM', 'FGA', 'FG%', '3PM', '3PA', '3P%',
           #'FTM', 'FTA', 'FT%', 'ORB', 'DRB', 'RPG', 'APG', 'SPG', 'BPG', 'TOV', 'PF']
#df_serbia_players = pd.DataFrame(data)

df_serbia_players['Pos'] = df_serbia_players['Pos'].replace(['PG', 'SG'], 'G')
df_serbia_players['Pos'] = df_serbia_players['Pos'].replace(['SF', 'PF'], 'F')


# Save DataFrame to a CSV file
df_serbia_players.to_csv('DataFiles/euroleague_stats_Serbia_player.csv', index=False)


# Display DataFrame
df_serbia_players
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
url = 'https://basketball.realgm.com/international/league/11/Israeli-BSL/players/1053/{year}'
# Send a GET request to the URL
response = requests.get(url)
# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')
    # Find the table containing player data
    table = soup.find('table', class_='tablesaw')
    # Extract column names
    columns = [th.text.strip() for th in table.find_all('th')]
    # Initialize data list to store parsed data
    data = []
    # Extract player data row by row
    for row in table.find_all('tr')[1:]:
        player_data = [td.text.strip() for td in row.find_all('td')]
        data.append(player_data)
    # Create DataFrame from the collected data
    df = pd.DataFrame(data, columns=columns)
    # Save DataFrame to a CSV file
    df.to_csv('DataFile/euroleague_stats_Israel_player.csv', index=False)
    # Display DataFrame
   df
```

```python
#renaming positions
df['Position'] = df['Position'].replace(['PG', 'SG'], 'G')
df['Position'] = df['Position'].replace(['SF', 'PF'], 'F')
```

```python
#calculation for 100 possessions
df.loc[:, 'GP':] = df.loc[:,'GP':].astype(float)
df.loc[:, 'MPG':] = df.loc[:,'MPG':].mul(1.25)
```
