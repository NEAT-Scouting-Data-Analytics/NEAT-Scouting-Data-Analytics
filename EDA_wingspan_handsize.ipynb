```python
df_cols_short = r['resultSets'][0]['headers']
```

```python
years = ['2000-01', '2001-02', '2002-03', '2003-04', '2004-05',
         '2005-06', '2006-07', '2007-08', '2008-09', '2009-10', '2010-11', '2011-12', '2012-13', '2013-14',
         '2014-15', '2015-16', '2016-17', '2017-18', '2018-19', '2019-20', '2020-21', '2021-22', '2022-23']

dataframes = []

for year in years:
    url = (f"https://stats.nba.com/stats/draftcombineplayeranthro?LeagueID=00&SeasonYear={year}")
    r = requests.get(url=url, headers=headers).json()
    temp_df = pd.DataFrame(r['resultSets'][0]['rowSet'], columns=df_cols_short)
    temp_df['SeasonYear'] = year
    dataframes.append(temp_df)
    df_anthro_players = pd.concat(dataframes, ignore_index=True)
```

```python
#Export dataframe 

df_anthro_players.to_csv('player_anthro.csv', index=False)
```

```python
# Assuming df_anthro_players is your DataFrame
# Count the number of null values in the 'HAND_WIDTH' and 'HAND_LENGTH' columns
null_count_hand_width = df_anthro_players['HAND_WIDTH'].isnull().sum()
null_count_hand_length = df_anthro_players['HAND_LENGTH'].isnull().sum()

print("Total number of null values in HAND_WIDTH column:", null_count_hand_width)
print("Total number of null values in HAND_LENGTH column:", null_count_hand_length)
```

```python
# Assuming df_anthro_players is your DataFrame
# Calculate the average of the 'HAND_WIDTH' and 'HAND_LENGTH' columns
average_hand_width = df_anthro_players['HAND_WIDTH'].mean()
average_hand_length = df_anthro_players['HAND_LENGTH'].mean()

# Fill null values with the average values
df_anthro_players['HAND_WIDTH_FILLED'] = df_anthro_players['HAND_WIDTH'].fillna(average_hand_width)
df_anthro_players['HAND_LENGTH_FILLED'] = df_anthro_players['HAND_LENGTH'].fillna(average_hand_length)

# Print out the new DataFrame with filled values
df_anthro_players.tail()
```
