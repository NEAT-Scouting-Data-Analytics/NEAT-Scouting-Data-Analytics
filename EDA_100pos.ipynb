```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
#Import dataframe

df_100pos = pd.read_csv('DataFiles/merged_df_100pos.csv', low_memory=False)
```

```python
# European countries in basketball sense
europe = ['Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', 'Czechia', 'Denmark',
  'Estonia', 'Finland', 'France', 'Germany', 'Greece', 'Hungary', 'Ireland',
  'Italy', 'Latvia', 'Lithuania', 'Luxembourg', 'Malta', 'Netherlands',
  'Poland', 'Portugal', 'Romania', 'Slovakia', 'Slovenia', 'Spain', 'Sweden', 
  'United Kingdom', 'Norway', 'Switzerland', 'Belarus', 'Russia', 'Ukraine',
  'Moldova', 'Georgia', 'Armenia', 'Azerbaijan', 'Turkey', 'Israel', 'Albania',
  'Macedonia', 'Serbia', 'Montenegro', 'Bosnia and Herzegovina', 'Iceland']
```

```python
# Function to filer on countries
def filter_country(df, column_name, country_list):
    return df[df[column_name].isin(country_list)]
```

```python
# New object for European players playing at F and F-C in regular season games:
europe_df = filter_country(df_100pos, 'COUNTRY', europe)
europe_df = europe_df.loc[europe_df['POSITION'].isin(['F','F-C']) & europe_df['Season'].isin(['Regular Season'])]
```

```python
europe_df.groupby('PLAYER_ID').unique().describe()
```

```python
# New object for USA players playing at F and F-C in regular season games:
US_df = filter_country(df_100pos, 'COUNTRY', ['USA'])
US_df = US_df.loc[US_df['POSITION'].isin(['F','F-C']) & US_df['Season'].isin(['Regular Season'])]
```

```python
# Number of positions per unique European player:
europe_df.groupby('POSITION')['PLAYER_ID'].nunique()
```

```python
# Number of positions per unique US player:
US_df.groupby('POSITION')['PLAYER_ID'].nunique()
```
