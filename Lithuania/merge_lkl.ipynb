```python
import pandas as pd
```

```python
#Import dataframe
df_stats = pd.read_csv('/Users/pandit1781/NEAT-Scouting-Data-Analytics/Lithuania/lkl_stats.csv', low_memory=False)
df_bio = pd.read_csv('/Users/pandit1781/NEAT-Scouting-Data-Analytics/Lithuania/lkl_playerbio.csv', low_memory=False)
```

```python
merged_df = pd.merge(df_stats, df_bio, on=['Player'], how='inner')
```

```python
# Players to whom more than 4 rows are assigned
merged_df[merged_df.groupby('Player').transform('size')>5]
```
