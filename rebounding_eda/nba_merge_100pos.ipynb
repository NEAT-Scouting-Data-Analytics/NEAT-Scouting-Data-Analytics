```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
#Import dataframes 

df_100pos = pd.read_csv('/Users/pandit1781/NEAT-Scouting-Data-Analytics/DataFiles/merged_df_100pos.csv', low_memory=False)
df_reb = pd.read_csv('/Users/pandit1781/NEAT-Scouting-Data-Analytics/rebounding_eda /nba_rebounding.csv', low_memory=False)
```

```python
# Clean 'df_reb' data set:

columns_to_delete = ['W', 'L', 'OREB_CHANCES', 'OREB_CHANCE_PCT', 'OREB_CHANCE_DEFER', 'OREB_CHANCE_PCT_ADJ', 'OREB_CHANCES', 'DREB_CHANCE_PCT', 
                     'DREB_CHANCE_DEFER', 'DREB_CHANCE_PCT_ADJ', 'AVG_OREB_DIST', 'AVG_DREB_DIST', 'REB_CHANCES', 'REB_CHANCE_PCT', 'REB_CHANCE_DEFER', 
                     'REB_CHANCE_PCT_ADJ', 'AVG_REB_DIST']

# Delete the specified columns
df_reb.drop(columns=columns_to_delete, inplace=True, errors='ignore')

# Rename
df_reb.rename(columns={'Season_type': 'Season'}, inplace=True)
```

```python
# Merge 'df_player_stats' and ‘df_bio‘ on common keys ‘PLAYER_ID‘ and ‘Year‘ in order to get 'COUNTRY' column.
df_reb_m = pd.merge(df_100pos, df_reb, on=['PLAYER_ID', 'Year', 'Season'], how='inner')   
```

```python
# Drop duplicate columns
columns_to_drop = ['PLAYER_NAME_y','TEAM_ID_y','TEAM_ABBREVIATION_y', 'GP_y', 'MIN_y', 'OREB_x', 'DREB_x', 'REB_x']
df_reb_m.drop(columns=columns_to_drop, inplace=True)
```

```python
# Drop '_x' at the end of column names 
for column in df_reb_m.columns: 
    if '_x' in column: 
        column_name = column.replace('_x', '')
        df_reb_m.rename(columns={column:column_name}, inplace=True)

# Change regular season name at 'Season' column  
df_reb_m['Season'] = df_reb_m['Season'].apply(lambda x:'Regular Season' if x =='Regular%20Season' else x)
```

```python
# Drop '_y' at the end of column names 
for column in df_reb_m.columns: 
    if '_y' in column: 
        column_name = column.replace('_y', '')
        df_reb_m.rename(columns={column:column_name}, inplace=True)
```
